## 获得列数据

### method1

```python
# 获得一行数据
d_col = f ['age']
# print(d_col)
# 获得多行数据\
cols = ['age','bmi','smoking_status']
d_cols = f[cols]
# print(f[cols].head())
print(d_cols.head())
```

### method2
loc[]

有两个输入参数，第一个指定行名，第二个指定列名。当只有一个参数时，默认是行名（即抽取整行），所有列都选中。

```python
# m = f.loc([行,列])
m= f.loc[0:5,'age']
m.head()
```

## 获得行数据

### method1

```python
d_row = f[0:5]
# type(d_row)
d_row.head()
```

### method2
data.iloc[行下标,列下标]

.iloc[]： 以二维矩阵的位置指标（即0,1,2……）作为参数。

```python
# d_row = f.iloc[1,['age']]
d_row = f.iloc[1:5, [0,1]]
d_row.head()
```
